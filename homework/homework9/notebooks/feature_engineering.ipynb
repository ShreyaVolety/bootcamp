```python
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys
import pandas as pd
import numpy as np


data_path = Path.cwd().parent / "data" / "processed" / "api_data_cleaned_merged.csv"
df = pd.read_csv(str(data_path),index_col=0)
df.index = df.Date
df = df.drop(columns=['Date'])
```

```python
returns_df = np.log(df / df.shift(1))
returns_df = returns_df.dropna()
```

```python
returns_df['CL_vol_5d'] = returns_df['CL=F'].rolling(5).std()
returns_df['CL_vol_21d'] = returns_df['CL=F'].rolling(21).std()
```

```python
returns_df['CL_mom_5d'] = returns_df['CL=F'].rolling(5).mean()
returns_df['CL_mom_21d'] = returns_df['CL=F'].rolling(21).mean()
```

```python
returns_df['UUP_lag1'] = returns_df['UUP'].shift(1)
returns_df['GLD_lag1'] = returns_df['GLD'].shift(1)
returns_df['VIX_lag1'] = returns_df['^VIX'].shift(1)
```

```python
save_path =  Path.cwd().parent / "data" / "processed" 
returns_df.to_csv(str(save_path)+"/engineered_features.csv")
```
